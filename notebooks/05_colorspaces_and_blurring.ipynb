```python
import os
import cv2
```

```python
file_path = os.path.join('..', 'saved_image', 'animal.jpg')
os.path.exists(file_path)
```

## All types of color conversion

https://docs.opencv.org/3.4.20/d8/d01/group__imgproc__color__conversions.html

```python
img = cv2.imread(file_path)
# changing the color space from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
cv2.imshow('img', img)
cv2.imshow('img_rgb', img_rgb)
cv2.imshow('img_gray', img_gray)
cv2.imshow('img_hsv', img_hsv)
cv2.waitKey(0)
```

## Blurring

```python
k_size = 7 # higher the number more the blurring
blurred_img = cv2.blur(img, (k_size, k_size)) # takes average value of that pixel surrounding
img_gaussian_blur = cv2.GaussianBlur(img, (k_size, k_size), 5) # takes weighted average value of that pixel surrounding
img_median_blur = cv2.medianBlur(img, k_size) # takes median value of that pixel surrounding
cv2.imshow('img', img)
cv2.imshow('img_blur', blurred_img)
cv2.imshow('img_gaussian_blur', img_gaussian_blur)
cv2.imshow('img_median_blur', img_median_blur)
cv2.waitKey(0)
```
