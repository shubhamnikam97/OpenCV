```python
import os
import cv2
```

```python
file_path = os.path.join(r'..\image\133935222844608753.jpg')
os.path.exists(file_path)
```

```python
img = cv2.imread(file_path)
print(img.shape)
```

```python
cv2.imshow('Image', img)
cv2.waitKey(0)
```

## Resize the image

```python
resize_image = cv2.resize(img, (1080, 640))
print(resize_image.shape)
cv2.imshow('Resize Image', resize_image)
cv2.waitKey(0)
```

## Crop the image

```python
cropped_image = resize_image[40:880,200:800]
print(cropped_image.shape)
cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)
```

```python
# writing the image
cv2.imwrite(os.path.join('..', 'saved_image', 'animal.jpg'), cropped_image)
```
