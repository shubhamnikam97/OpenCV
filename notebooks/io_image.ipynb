```python
import os
import cv2
import numpy as np
```

```python
os.path.exists(r'..\image\133935222844608753.jpg')
```

```python
image_path = os.path.join('..', 'image', '133935222844608753.jpg')
os.path.exists(image_path)
```

```python
# reading the image (in numpy format)
img = cv2.imread(image_path)
img.shape # (height, width, channels)
```

```python
# writing the image
cv2.imwrite(os.path.join('..', 'saved_image', '133935222844608753.jpg'), img)
```

```python
# displaying the image
cv2.imshow('Image', img)
cv2.waitKey(0) # wait until a key is pressed
cv2.waitKey(5000) # wait for 5 seconds (5000 milliseconds)
```
