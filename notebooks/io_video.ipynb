```python
import os
import cv2
import numpy as np
```

```python
video_path = os.path.join('..', 'video', 'testing.mp4')
os.path.exists(video_path)
```

```python
# read video
video = cv2.VideoCapture(video_path)
```

```python
# visulize video
ret = True
while ret:
    ret, frame = video.read()

    if ret:
        cv2.imshow('frame', frame)
        cv2.waitKey(40)
```

```python
# release memory allocated to video and close all windows
video.release()
cv2.destroyAllWindows()
```
